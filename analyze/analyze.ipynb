## Decomposition

The following decomposition, satisfy Boyce-codd and the third normal form at the same time.

### Relational schema

**event** (<ins>id</ins> | url, title, lead_text, description, date_start, date_end, occurrence, cover_url, cover_alt, cover_credit, tags, address_name, address_street, address_zipcode, address_city, price_type, price_detail, access_type, access_link, access_link_text, updated_at, image_couverture, programs, address_url, address_url_text, address_text, title_event, audience, contact_id)

**contact**(<ins>contact_id</ins>|contact_url, contact_phone, contact_mail, contact_facebook, contact_twitter)

**geographic_correspondance**(<ins>address_street, address_zipcode, address_city </ins> | lat, lon)

**transport**(<ins>lat, lon</ins>| type, line, station)

**location_information**(<ins>address_name, address_street, address_zipcode, address_city</ins>| pmr, blind, deaf)

**sub_events**(<ins>children_event_id</ins>, parent_event_id)
